```python
# Load ruamel or pyyaml as needed
try:
    import ruamel.yaml
    yaml = ruamel.yaml.YAML(typ='unsafe', pure=True)
    useruamel=True
    loaderkwargs = {'Loader':yaml.RoundTripLoader}
    dumperkwargs = {'Dumper':yaml.RoundTripDumper, 'indent':4, 'default_flow_style':False}
    Loader=yaml.load
except:
    import yaml as yaml
    useruamel=False
    loaderkwargs = {}
    dumperkwargs = {'default_flow_style':False }
    Loader=yaml.safe_load
    
```

```python
#For the King Planes Turbines - OpenFast data
of_dir = "/lustre/orion/cfd162/world-shared/lcheung/AWAKEN.cli187/Benchmark1/Phase1/BM1_FarmRunProd1.summit/"
run_ppeng = True
```

```python
yamlstring="""

globalattributes:
  verbose: False
  udfmodules: []
  executeorder:
  - openfast
  #- plotcsv

#trange: &trange [65, 125]
  
openfast:
# For FSI case
- name: []
  filename: []
  vars:
  - Time
#  - Wind1VelX
#  - Wind1VelY
#  - Wind1VelZ
  - GenPwr
  - RotSpeed
  - RotThrust
  output_dir: OpenFAST_KP
  csv:  # Store information to CSV files
    individual_files: False
"""
f = io.StringIO(yamlstring)
yamldict = Loader(f, **loaderkwargs)
```

```python
N_KP = 88

# Fill in the names and filenames
KPturbnames = ['T%02i'%(i+1) for i in range(N_KP)]
KPturbdirs = [of_dir + '/WIT_USKPL_SS001_WT%03i_OpenFAST_NREL2p8-127_AWAKEN/NREL-2p8-127.T%i.out'%(i+1,i) for i in range(N_KP)]
KPturbdirs[0] = KPturbdirs[0].replace('.T0.out', '.out')
yamldict['openfast'][0]['name']     = KPturbnames 
yamldict['openfast'][0]['filename'] = KPturbdirs
```

```python
# Run the driver
if run_ppeng:
    ppeng.driver(yamldict, verbose=True)
```
