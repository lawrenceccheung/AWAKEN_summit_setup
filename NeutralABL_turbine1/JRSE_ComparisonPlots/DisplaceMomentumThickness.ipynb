# Compare displacement/momentum thickness

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

from matplotlib.lines import Line2D
```

```python
# Put the list of all cases in here
cases = [
    # AMR-Wind data defs
    {'label':'AMR-Wind',
     'datadir':'../AMR.5kmX5km_turbine1/post_processing/data',
     'prefix':'AMRWIND',
     'fname':'{prefix}_DeltaThetaH_Thickness.dat',
     'linestyle':{'linestyle':'-', 'lw':2},
    },
    
    # Nalu data defs
    {'label':'NALUWIND',
     'datadir':'../NALU.5kmX5km_turbine1/postprocessing/data',
     'prefix':'NALUWIND',
     'fname':'{prefix}_DeltaThetaH_Thickness.dat',
     'linestyle':{'linestyle':'--', 'lw':2},
    },

]
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,2.5), dpi=125)
for icase, case in enumerate(cases):
    label   = case['label']
    prefix  = case['prefix']
    datadir = case['datadir']
    lstyle  = case['linestyle']
    fname   = case['fname'] #f'{prefix}_HH_VEL_PROFILE_x_{xD:06.2f}D.dat'
    dat=np.loadtxt(datadir+'/'+fname.format(prefix=prefix))
    ax.plot(dat[:,0], dat[:,1],  label=prefix, **lstyle) # color=cycle[ix],

ax.legend()
ax.set_ylabel('$\delta/A$')
ax.set_xlabel('x/D')
ax.set_title('Displacement thickness')
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,2.5), dpi=125)
for icase, case in enumerate(cases):
    label   = case['label']
    prefix  = case['prefix']
    datadir = case['datadir']
    lstyle  = case['linestyle']
    fname   = case['fname'] 
    dat=np.loadtxt(datadir+'/'+fname.format(prefix=prefix))
    ax.plot(dat[:,0], dat[:,2],  label=prefix, **lstyle) # color=cycle[ix],

ax.legend()
ax.set_ylabel(r'$\theta/A$')
ax.set_xlabel('x/D')
ax.set_title('Momentum thickness')
```
