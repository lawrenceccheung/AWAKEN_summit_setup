```python
def getTurbSubset(csvfile, suffix):
    # Load the data from the database of all turbines
    allturbsdf = pd.read_csv(csvfile, low_memory=False)
    # Get just the King Plains turbines
    subset = [] 
    for x in allturbsdf.iterrows():
        if x[1]['# name'].endswith(suffix):
            subset.append(x[1])
    subsetdf = pd.DataFrame(subset[:])
    return subsetdf

def getTurbSubsetXY(csvfile, suffix, xlim=[-np.finfo(np.float64).max, np.finfo(np.float64).max], ylim=[-np.finfo(np.float64).max, np.finfo(np.float64).max]):
    # Load the data from the database of all turbines
    allturbsdf = pd.read_csv(csvfile, low_memory=False)
    # Get just the King Plains turbines
    subset = [] 
    for x in allturbsdf.iterrows():
        xy = [float(x[1][' x']), float(x[1][' y'])]
        if x[1]['# name'].endswith(suffix) and ((xlim[0] <= xy[0]) and (xy[0]<=xlim[1]) and (ylim[0]<=xy[1]) and (xy[1]<=ylim[1])):
            #print(x[1])
            subset.append(x[1])
    subsetdf = pd.DataFrame(subset[:])
    return subsetdf

getturbnames = lambda df: [x[1]['# name'] for x in df.iterrows()]

def setfigtextsize(ax, fsize):
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label, ax.yaxis.get_offset_text()] + ax.get_xticklabels() + ax.get_yticklabels() ):
        item.set_fontsize(fsize)
    return
        
# Helper to read the OpenFAST out file
readoutfile = lambda f: pd.read_csv(f,sep='\s+',skiprows=(0,1,2,3,4,5,7), header=(0),skipinitialspace=True)
```

## Plot OpenFAST power

Note: The full OpenFAST output file is too big to load N number of times.  Use the `extractpower.sh` script to pull out just the power variable plot.

```python
with open(setupdir+"/post_processing/OFturbpower/extractpower.sh") as f: print(f.read())
```

```python
def avgturbpower(turbs, turbdirtemplate, OFturbpattern):
    avgpower = None
    for turb in turbs:
        turbdir = turbdirtemplate%turb #setupdir+'/'+OFoutdir+'/'+turb+OFdirsuffix
        searchpattern = turbdir+'/'+OFturbpattern
        outfilelist=glob.glob(searchpattern)
        OFout = readoutfile(outfilelist[0])
        time = OFout['Time']
        if avgpower is None:
            avgpower = OFout['GenPwr'][:]
        else:
            avgpower += OFout['GenPwr'][:]
        del OFout
    avgpower /= len(turbs)
    return time, avgpower
```

```python
time, frontrow_avgpower = avgturbpower(classifyKPturbines.turbinecategorylist['frontrow'], setupdir+'/'+OFoutdir+'/%s'+OFdirsuffix, '*')
time, backrow_avgpower  = avgturbpower(classifyKPturbines.turbinecategorylist['backrow'], setupdir+'/'+OFoutdir+'/%s'+OFdirsuffix, '*')
time, midrow_avgpower   = avgturbpower(classifyKPturbines.turbinecategorylist['midrow'], setupdir+'/'+OFoutdir+'/%s'+OFdirsuffix, '*')
time, AFwaked_avgpower  = avgturbpower(classifyKPturbines.turbinecategorylist['AFwaked'], setupdir+'/'+OFoutdir+'/%s'+OFdirsuffix, '*')
time, all_avgpower      = avgturbpower(classifyKPturbines.allturbines, setupdir+'/'+OFoutdir+'/%s'+OFdirsuffix, '*')
```

```python
turblist = getturbnames(getTurbSubset(setupdir+'/'+summitcsv, '-KP')) #getturbnames(getTurbSubsetXY(setupdir+'/'+summitcsv, '-KP', ylim=[0, 4026000]))
#avgpower = None
toffset  = 16125
#fig, ax = plt.subplots(1,1,figsize=(4,2.5), dpi=125)
fig, ax = plt.subplots(1,1,figsize=(8,6), dpi=125)
for turb in turblist:
    turbdir = setupdir+'/'+OFoutdir+'/'+turb+OFdirsuffix
    searchpattern = turbdir+'/'+OFturbpattern
    outfilelist=glob.glob(searchpattern)
    #print(searchpattern, outfilelist)
    #print(outfilelist[0])
    OFout = readoutfile(outfilelist[0])
    time = OFout['Time']
    #if avgpower is None:
    #    avgpower = OFout['GenPwr'][:]
    #else:
    #    avgpower += OFout['GenPwr'][:]
    ax.plot(time+toffset, OFout['GenPwr'], color='lightgray', lw=0.25, alpha=0.5, label='')
    del OFout
#print(len(turblist))
#avgpower /= len(turblist)


ax.plot(time+toffset, backrow_avgpower, color='b', lw=1, label='Back row Avg. power')
ax.plot(time+toffset, midrow_avgpower, color='y', lw=1, label='Mid row Avg. power')
ax.plot(time+toffset, frontrow_avgpower, color='r', lw=1, label='Front row Avg. power')
ax.plot(time+toffset, AFwaked_avgpower, color='pink', lw=1, label='AF Waked Avg. power')

ax.plot(time+toffset, all_avgpower, color='k', lw=2, label='Overall Avg. power')

ax.set_title('King Plains turbine power')
ax.set_ylim([0, 2700])
ax.set_xlim([16125, 17300])
ax.set_xlabel('Time [s]')
ax.set_ylabel('Gen. Power [kW]')
ax.legend(prop={'size':fontsize}, loc='lower left')
ax.grid(ls=':', color='b')
setfigtextsize(ax,fontsize)
```

```python
def avgpower(tvec, pvec, tavg):
    filter = (tavg[0]<= tvec) & (tvec <= tavg[1])
    tfilter = np.array(tvec[filter])
    pfilter = np.array(pvec[filter])
    return np.mean(pfilter)
```

```python
nodraftpower = avgpower(time+toffset, nodraft_avgpower, [16125+300, 16125+960])
updraftpower = avgpower(time+toffset, updraft_avgpower, [16125+300, 16125+960])
downdraftpower = avgpower(time+toffset, downdraft_avgpower, [16125+300, 16125+960])

print('downdraft: %f'%downdraftpower)
print('updraft:   %f'%updraftpower)
print('nodraft:   %f'%nodraftpower)
```
